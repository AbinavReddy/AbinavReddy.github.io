## What Does ONE Plot Show?

Let's focus on a single plot, for example: **Assault vs Larceny/Theft**

```
        Larceny/Theft (y-axis)
        ↑
    800 │                    •
        │              •
    600 │         •        •
        │    •         •
    400 │  •      •
        │•    •
    200 │•
        │
      0 └─────────────────────────→
        0   200  400  600  800
              Assault (x-axis)
```

### What Each Point Represents:

**Each dot = 1 hour of the week (168 total dots)**

For example:
- One dot might be "Sunday 2am"
- X-coordinate = How many Assaults happened during Sunday 2am (across all years)
- Y-coordinate = How many Larceny/Thefts happened during Sunday 2am (across all years)

Another dot might be "Friday 11pm":
- X-coordinate = Assaults during Friday 11pm
- Y-coordinate = Larceny/Thefts during Friday 11pm

## What Patterns Tell Us

### Pattern 1: Strong Positive Correlation (Diagonal Line)

```
Larceny/Theft
    ↑
800 │                    •
    │                 •
600 │              •
    │           •
400 │        •
    │     •
200 │  •
    │•
  0 └─────────────────────→
    0   200  400  600  800
         Assault
```

**What this means:**
- When Assault is HIGH, Larceny/Theft is also HIGH
- When Assault is LOW, Larceny/Theft is also LOW
- **Interpretation:** These crimes follow similar weekly patterns
  - Maybe both peak on Friday/Saturday nights
  - Maybe both are low on Tuesday mornings
  - They're influenced by similar factors (nightlife, crowds, etc.)

### Pattern 2: No Correlation (Random Scatter)

```
Embezzlement
    ↑
 50 │  •     •
    │    •       •
 40 │       •  •
    │  •   •
 30 │     •    •
    │ •      •
 20 │   •  •
    │ •
  0 └─────────────────────→
    0   200  400  600  800
       Liquor Laws
```

**What this means:**
- No clear pattern
- When Liquor Laws is HIGH, Embezzlement could be HIGH or LOW
- **Interpretation:** These crimes follow DIFFERENT weekly patterns
  - Liquor Laws violations peak on weekend nights (bar enforcement)
  - Embezzlement is discovered during business hours (Monday-Friday)
  - They're influenced by completely different factors

## The Color Coding (Exercise 3.2)

In the enhanced version, each point is colored by its hour of the week:

```
🔴 Red points   = Sunday hours
🟠 Orange       = Monday hours  
🟡 Yellow       = Tuesday hours
🟢 Green        = Wednesday hours
🔵 Blue         = Thursday-Saturday hours
```

**Why is this useful?**

You can see if certain times of week cluster together:
- If all the red points (Sunday) are in the bottom-left → Both crimes are LOW on Sundays
- If all the blue points (Saturday) are in the top-right → Both crimes are HIGH on Saturdays

## The Regression Line (Exercise 3.2)

The black dashed line is the "best fit" line through the points:

```
Larceny/Theft
    ↑
800 │                    •
    │                 • /
600 │              • /
    │           • /
400 │        • /
    │     • /
200 │  • /
    │• /
  0 └─────────────────────→
    0   200  400  600  800
         Assault
         
Line equation: y = 1.2x + 50
```

**What the line tells us:**
- Slope = 1.2 means: For every 1 additional Assault, we expect 1.2 additional Larceny/Thefts
- The line helps us see the overall trend even if points are scattered

## R² Value (Exercise 3.3)

The R² number tells us **how well the line fits the data**:

```
R² = 0.95  →  Excellent fit (points very close to line)
R² = 0.70  →  Good fit (points somewhat close to line)
R² = 0.30  →  Weak fit (points scattered)
R² = 0.05  →  No fit (random scatter)
```

**Interpretation:**
- R² = 0.85 means "85% of the variation in Crime Y can be explained by Crime X"
- High R² = Strong correlation = Similar temporal patterns
- Low R² = Weak correlation = Different temporal patterns

## Putting It All Together: What Are We Looking For?

When you look at the 6x6 grid, you're asking:

### Question 1: Which crime pairs have similar weekly rhythms?
**Look for:** Diagonal lines, high R² values
**Example:** Assault vs Disorderly Conduct (both peak on weekend nights)

### Question 2: Which crime pairs have different weekly rhythms?
**Look for:** Random scatter, low R² values
**Example:** Embezzlement vs Liquor Laws (business hours vs nightlife)

### Question 3: Why do some crimes correlate?
**Think about:**
- Do they happen in similar contexts? (street crimes, nightlife, etc.)
- Do they share opportunity structures?
- Are they influenced by the same factors? (alcohol, crowds, darkness)

### Question 4: Why don't some crimes correlate?
**Think about:**
- Different contexts (white-collar vs street crime)
- Different reporting mechanisms (immediate vs delayed)
- Different times of day (business hours vs nightlife)
- Different victim/offender profiles

## Still Confused? Let's Use an Analogy

Think of it like comparing **ice cream sales** vs **sunglasses sales**:

```
Sunglasses Sales
    ↑
    │                    • (Summer)
    │                 •
    │              •
    │           •
    │        •
    │     •
    │  • (Winter)
    │•
    └─────────────────────→
         Ice Cream Sales
```

**What we see:**
- Diagonal line (strong correlation)
- When ice cream sales are HIGH (summer), sunglasses sales are HIGH
- When ice cream sales are LOW (winter), sunglasses sales are LOW

**Why?**
- Both are driven by the same factor: WEATHER
- Sunny, hot days → people buy both
- Cold, dark days → people buy neither

**Same logic for crimes:**
- Assault and Disorderly Conduct both driven by: NIGHTLIFE
- Weekend nights → both are high
- Tuesday mornings → both are low
