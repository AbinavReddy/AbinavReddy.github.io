# Part 1: Predictive Policing—A Cautionary Tale

For years, "predictive policing" was pitched as the future of law enforcement. The idea was simple: use data and algorithms to predict where crimes would occur, then deploy police resources accordingly. It connected to popular culture through shows like [NUMB3RS](https://en.wikipedia.org/wiki/Numbers_(TV_series)) and films like [Minority Report](https://www.imdb.com/title/tt0181689/). Companies like PredPol and Palantir built businesses selling these tools to police departments.

But the story didn't end the way proponents hoped.

**By the early 2020s, many cities abandoned these systems:**
- The **LAPD dropped PredPol in 2020** after audits showed it failed to reduce crime and disproportionately targeted Black and Latino neighborhoods
- **Santa Cruz became the first U.S. city to ban predictive policing** entirely
- Academic research revealed that these systems often **amplified existing biases** rather than providing objective predictions

This makes predictive policing a perfect case study for our class. It shows how data analysis and visualization can be powerful—but also how they can go wrong when we don't think carefully about bias, feedback loops, and the human consequences of algorithmic decisions.

### **1. What did ProPublica find about COMPAS’s accuracy for Black vs. white defendants?**

ProPublica’s investigation of the COMPAS recidivism risk algorithm found that while the **overall predictive accuracy** for future crime was similar across races (about 61 % correct), the **types of errors differed markedly** between Black and white defendants. Black defendants who did **not** reoffend were **almost twice as likely** as white defendants to be labeled *high risk* (false positives). Conversely, white defendants who *did* reoffend were more often labeled *low risk* compared to Black reoffenders (false negatives). Even after controlling for age, prior crimes, and actual recidivism, Black defendants were significantly more likely to receive higher risk scores than white defendants. :contentReference[oaicite:0]{index=0}

### **2. What is “dirty data,” and why can’t we just fix the algorithm? Do you agree with this argument?**

**“Dirty data”** refers to police records and datasets that are shaped by **biased, discriminatory, or unlawful policing practices**. This includes patterns of over-policing in certain communities, racially disparate stops and arrests, and other civil rights violations that distort the recorded data used to train predictive systems. Because predictive algorithms learn patterns from this historical data, they incorporate not only patterns of crime but also patterns of policing behavior — including biases and errors embedded in how the data was generated. :contentReference[oaicite:1]{index=1}

Fixing the algorithm alone does **not** solve the problem if the input data itself reflects biased enforcement practices. Any model, regardless of mathematical sophistication, will reproduce patterns present in its training data. Algorithms cannot determine whether the correlations they rely on arise from actual criminal behavior or from biased policing; they simply extrapolate from what is in the data. Therefore, efforts that focus only on algorithmic adjustments without addressing **the underlying data generation process** will continue to produce flawed and potentially discriminatory outputs.

I agree with this argument because biased inputs limit the ability of algorithmic adjustments to produce fair or unbiased predictions — systemic biases embedded in the data will continue to influence outcomes regardless of changes to model structure or fairness constraints.

### **3. Why did cities like Los Angeles and Santa Cruz decide to abandon predictive policing?**

Cities such as Los Angeles and Santa Cruz discontinued predictive policing because of concerns that these systems **reinforced existing racial and geographic disparities in policing** rather than mitigating them. Predictive policing tools often rely on historical police data, which may be tainted by unequal policing practices. Using such data to direct future enforcement can create **feedback loops**: areas already heavily policed generate more data and are then predicted to be high-crime zones, leading to increased police presence and further data collection — irrespective of actual crime levels. :contentReference[oaicite:2]{index=2}

In jurisdictions with documented histories of biased policing, this risk is particularly acute and raises concerns about civil rights violations and community trust. The lack of transparency in how these systems operate and the potential for perpetuating harm contributed to the decision to abandon them. :contentReference[oaicite:3]{index=3}

```python
import pandas as pd

# Load the data
df = pd.read_csv(r"C:\Users\abina\Downloads\Police_Department_Incident_Reports__2018_to_Present_20260203.csv")

# Convert date columns to datetime (let pandas infer format, suppress warning)
df['incident_date'] = pd.to_datetime(df['Incident Date'])

# Extract year
df['year'] = df['incident_date'].dt.year

# Check the original date range
print("="*60)
print("ORIGINAL DATA")
print("="*60)
print(f"Date range: {df['incident_date'].min().date()} to {df['incident_date'].max().date()}")
print(f"Total rows: {len(df):,}")
print(f"Years in data: {sorted(df['year'].unique())}")

# Keep only complete years (2018 through 2025)
# Since today is Feb 3, 2026, we drop 2026 (incomplete)
# Data starts Jan 1, 2018, so 2018 is complete
df_clean = df[(df['year'] >= 2018) & (df['year'] <= 2025)].copy()

# Keep relevant columns: incident category, date/time, location (coords + police district)
columns_to_keep = [
    'Incident Datetime', 'Incident Date', 'Incident Time', 'year',
    'Incident Category', 'Incident Subcategory', 'Incident Description',
    'Police District', 'Latitude', 'Longitude', 'Resolution'
]
df_clean = df_clean[[col for col in columns_to_keep if col in df_clean.columns]]

# Report the results as requested
print("\n" + "="*60)
print("CLEANED DATASET SUMMARY (Exercise 2.2)")
print("="*60)
print(f"\n1. Total number of incidents: {len(df_clean):,}")
print(f"\n2. Date range covered: {df_clean['Incident Date'].min()} to {df_clean['Incident Date'].max()}")
print(f"\n3. Number of incident categories: {df_clean['Incident Category'].nunique()}")

# Show the categories
print("\n" + "-"*60)
print("Incident Categories (sorted by frequency):")
print("-"*60)
print(df_clean['Incident Category'].value_counts())
```

```python
# Exercise 3.2: Visualize crime categories
import matplotlib.pyplot as plt

# Filter out uninformative categories and get top 20
uninformative = ['Other Miscellaneous', 'Other Offenses', 'Other', 'Non-Criminal']
category_counts_filtered = category_counts[~category_counts.index.isin(uninformative)]
top_20 = category_counts_filtered.head(20)

# Create horizontal bar chart
fig, ax = plt.subplots(figsize=(12, 10))

# Plot bars (reversed so most common is at top)
bars = ax.barh(range(len(top_20)), top_20.values[::-1], color='steelblue', edgecolor='white')

# Set y-tick labels
ax.set_yticks(range(len(top_20)))
ax.set_yticklabels(top_20.index[::-1], fontsize=11)

# Labels and title
ax.set_xlabel('Number of Incidents', fontsize=12)
ax.set_ylabel('Incident Category', fontsize=12)
ax.set_title('Top 20 Crime Categories in San Francisco (2018-2025)\n(Excluding "Other" categories)', fontsize=14, fontweight='bold')

# Add value labels on bars
for i, (bar, val) in enumerate(zip(bars, top_20.values[::-1])):
    ax.text(val + 1000, bar.get_y() + bar.get_height()/2, f'{val:,}', 
            va='center', fontsize=9, color='black')

# Add gridlines
ax.xaxis.grid(True, linestyle='--', alpha=0.7)
ax.set_axisbelow(True)

# Format x-axis with thousands separator
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.tight_layout()
plt.show()

# Also show log scale version
fig, ax = plt.subplots(figsize=(12, 10))
bars = ax.barh(range(len(top_20)), top_20.values[::-1], color='coral', edgecolor='white')
ax.set_yticks(range(len(top_20)))
ax.set_yticklabels(top_20.index[::-1], fontsize=11)
ax.set_xlabel('Number of Incidents (log scale)', fontsize=12)
ax.set_ylabel('Incident Category', fontsize=12)
ax.set_title('Top 20 Crime Categories (Log Scale)\nShows relative differences more clearly', fontsize=14, fontweight='bold')
ax.set_xscale('log')
ax.xaxis.grid(True, linestyle='--', alpha=0.7)
ax.set_axisbelow(True)
plt.tight_layout()
plt.show()
```

### **Exercise 4.3: Interpretation of Focus Crime Trends**

Looking at the 4x4 grid of focus crime plots, here are **three interesting patterns**:

---

#### **1. COVID-19 Had Vastly Different Effects on Different Crime Types**

The pandemic year (2020, shown in red) reveals dramatically different impacts:

- **Crimes that DROPPED sharply in 2020:**
  - **Larceny Theft** (-37%) — Fewer people in stores/public spaces to steal from
  - **Prostitution** (-79%) — Dramatic decline due to social distancing
  - **Drug Offense** (-29%) — Less street activity, fewer enforcement opportunities
  - **Disorderly Conduct** (-35%) — Fewer public gatherings

- **Crimes that INCREASED in 2020:**
  - **Burglary** (+53%) — Empty commercial buildings became targets
  - **Motor Vehicle Theft** (+40%) — Parked cars left unattended for longer
  - **Vandalism** (+68%) — Boarded-up businesses, empty streets
  - **Arson** (+37%) — Related to protests and vacant buildings

This divergence makes sense: crimes requiring human interaction dropped, while crimes targeting unattended property surged.

---

#### **2. Some Crimes Show Long-Term Declining Trends (Beyond COVID)**

Several crimes show sustained decline that started before 2020 and continued after:

- **Prostitution** dropped from ~300 incidents (2018) to ~100 (2025) — an ongoing 70% decline
- **Sex Offense** decreased from ~175 (2018) to ~50 (2025)
- **Robbery** declined steadily from ~3,200 (2018) to ~2,400 (2025)

These trends may reflect:
- Changes in policing priorities
- Social/economic factors
- Shifts in how crimes are reported or categorized

---

#### **3. Drug Offense Shows a Unique "V-Shape" Recovery Pattern**

Drug offenses tell a particularly interesting story:
- **2018-2020:** Steady decline culminating in a pandemic low (~2,000)
- **2021-2025:** Sharp rebound, now exceeding pre-pandemic levels (~6,500)

This pattern likely reflects:
- **Enforcement-driven data:** Drug arrests depend heavily on police presence and priorities
- **Policy changes:** San Francisco's approach to drug enforcement may have shifted
- **The opioid/fentanyl crisis:** Worsening substance abuse issues post-pandemic

This is exactly the kind of pattern the "dirty data" paper warned about — we can't tell if drug *use* actually changed, or if what changed was *enforcement patterns*.

---

#### **Key Takeaway**

Different crimes behave very differently! Aggregating all crimes into a single trend would mask these important variations. This is why examining crime types individually is essential for understanding what's actually happening in the data.
