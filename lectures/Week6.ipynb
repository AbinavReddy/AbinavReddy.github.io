> ### Data prep
>
> A key step is to set up the data right. I recommend the following workflow:
>
> 1. Take the data for the period you want to plot, group it by hour-of-the-day for each of your Personal Focus Crimes. We'll use the entire period by default, but you may use another period if you'd like.
> 2. We want to compare *shapes* of distributions across crime types, not absolute numbers, so we'll work on **normalized data**:
>    * To normalize within a crime category, divide the count for each hour by the total number of that crime type. (For example, in the `Drug Offense` category, take the count of drug offenses in the midnight-to-1am hour and divide by the total number of drug offenses across all hours, etc)
>    * Note that when you normalize the data, you can interpret your data as a probability distribution, where each hour shows the probability that a crime will happen at that time.
>    * Your life will be easiest if you organize your data in **long format** — a DataFrame with columns like `Hour`, `CrimeType`, and `NormalizedCount`. If you have it in wide format (one column per crime type), look into [`pd.melt()`](https://pandas.pydata.org/docs/reference/api/pandas.melt.html) to reshape it.
>
> ### Creating the plot
>
> You'll use `px.bar` from Plotly Express — the same library you used for choropleth maps in Week 5. A few things to think about:
>
> * With many crime types, placing bars side-by-side for each hour produces tiny, unreadable bars. Look into Plotly's different [`barmode`](https://plotly.com/python/bar-charts/) options for a better approach. Adding some transparency will help when bars overlap.
> * The plot is more useful if it starts "empty" (all data/traces hidden) so the user can build up comparisons by clicking legend items one at a time. Look into Plotly's `visible` trace property to achieve this.
> * If all traces start hidden, Plotly has no data to auto-scale from, so your axes will not be meaningful. You'll need to set `range_x` and `range_y` explicitly. Think about what sensible values would be for 24 hours of data with normalized counts.
> * Place the legend where it doesn't obscure the data or the axis tick values. <font color=gray>It's OK to ask your LLM for help with legend placement.</font>
>

````python
import pandas as pd
import plotly.express as px

# Load the data
df = pd.read_csv('../files/merged_crime_data_2003_2025_clean.csv', parse_dates=['Incident_Date'])

# Extract hour from Incident_Time
time_parsed = pd.to_datetime(df['Incident_Time'], format='%H:%M', errors='coerce')
df['Hour'] = time_parsed.dt.hour

# Extract year and keep only whole years (2003-2024)
df['Year'] = df['Incident_Date'].dt.year
df = df[(df['Year'] >= 2003) & (df['Year'] <= 2024)]

# Personal Focus Crimes
focus_crimes = [
    'Assault', 'Burglary', 'Larceny/Theft', 'Robbery', 'Arson',
    'Vehicle Theft', 'Embezzlement', 'Fraud', 'Liquor Laws',
    'Suicide', 'Missing Person', 'Disorderly Conduct'
]

# Filter to focus crimes
df_focus = df[df['Unified_Category'].isin(focus_crimes)].copy()

# Group by Hour and CrimeType
hourly_counts = df_focus.groupby(['Hour', 'Unified_Category']).size().reset_index(name='Count')

# Normalize: divide each hour's count by the total for that crime type
totals = hourly_counts.groupby('Unified_Category')['Count'].transform('sum')
hourly_counts['NormalizedCount'] = hourly_counts['Count'] / totals

# Rename for plotting
hourly_counts = hourly_counts.rename(columns={'Unified_Category': 'CrimeType'})

# Create interactive bar chart
fig = px.bar(
    hourly_counts,
    x='Hour',
    y='NormalizedCount',
    color='CrimeType',
    barmode='overlay',
    opacity=0.6,
    title='Hourly Distribution of Focus Crimes (Normalized)',
    labels={'NormalizedCount': 'Probability', 'Hour': 'Hour of Day', 'CrimeType': 'Crime Type'},
    range_x=[-0.5, 23.5],
    range_y=[0, 0.15]
)

# Start with all traces hidden so user can toggle them on
fig.update_traces(visible='legendonly')

# Place legend to the right, outside the plot area
fig.update_layout(
    legend=dict(
        title='Crime Type',
        orientation='v',
        yanchor='top',
        y=1,
        xanchor='left',
        x=1.02
    ),
    xaxis=dict(dtick=1),
    width=900,
    height=500,
    margin=dict(r=200)
)

fig.show()
````

````python
import pandas as pd
import plotly.express as px

# Load the data
df = pd.read_csv('../files/merged_crime_data_2003_2025_clean.csv', parse_dates=['Incident_Date'])

# Extract hour and year
time_parsed = pd.to_datetime(df['Incident_Time'], format='%H:%M', errors='coerce')
df['Hour'] = time_parsed.dt.hour
df['Year'] = df['Incident_Date'].dt.year
df = df[(df['Year'] >= 2003) & (df['Year'] <= 2024)]

# Personal Focus Crimes
focus_crimes = [
    'Assault', 'Burglary', 'Larceny/Theft', 'Robbery', 'Arson',
    'Vehicle Theft', 'Embezzlement', 'Fraud', 'Liquor Laws',
    'Suicide', 'Missing Person', 'Disorderly Conduct'
]

df_focus = df[df['Unified_Category'].isin(focus_crimes)].copy()

# --- Check yearly counts to spot precipitous drops ---
yearly_totals = df_focus.groupby(['Year', 'Unified_Category']).size().reset_index(name='Count')
print("Yearly totals (last few years) for each crime type:")
pivot = yearly_totals.pivot(index='Year', columns='Unified_Category', values='Count').fillna(0)
print(pivot.tail(10).to_string())

# Group by Year, Hour, and CrimeType
hourly_yearly = df_focus.groupby(['Year', 'Hour', 'Unified_Category']).size().reset_index(name='Count')

# Create a full index of all (Year, Hour, CrimeType) combinations so no gaps exist
all_years = sorted(hourly_yearly['Year'].unique())
all_hours = range(24)
all_crimes = sorted(hourly_yearly['Unified_Category'].unique())
full_idx = pd.MultiIndex.from_product([all_years, all_hours, all_crimes],
                                       names=['Year', 'Hour', 'Unified_Category'])
hourly_yearly = hourly_yearly.set_index(['Year', 'Hour', 'Unified_Category']).reindex(full_idx, fill_value=0).reset_index()

# Normalize within each (Year, CrimeType) pair
totals = hourly_yearly.groupby(['Year', 'Unified_Category'])['Count'].transform('sum')
# Avoid division by zero for crime types with 0 total in a given year
hourly_yearly['NormalizedCount'] = hourly_yearly['Count'] / totals.replace(0, 1)

# Rename for plotting
hourly_yearly = hourly_yearly.rename(columns={'Unified_Category': 'CrimeType'})

# Ensure Year is int for clean animation labels
hourly_yearly['Year'] = hourly_yearly['Year'].astype(int)

# --- Animated LINE chart ---
fig_line = px.line(
    hourly_yearly,
    x='Hour',
    y='NormalizedCount',
    color='CrimeType',
    animation_frame='Year',
    title='Hourly Crime Distribution by Year (Animated Line Chart)',
    labels={'NormalizedCount': 'Probability', 'Hour': 'Hour of Day', 'CrimeType': 'Crime Type'},
    range_x=[-0.5, 23.5],
    range_y=[0, 0.2]
)

# Start with all traces hidden — only on base traces, NOT on frames.
# Setting visible on frame traces would override legend toggling during animation.
fig_line.update_traces(visible='legendonly')

fig_line.update_layout(
    legend=dict(
        title='Crime Type',
        orientation='v',
        yanchor='top', y=1,
        xanchor='left', x=1.02
    ),
    xaxis=dict(dtick=1),
    width=950,
    height=550,
    margin=dict(r=200)
)

fig_line.show()
````

````python
# --- Animated BAR chart ---
fig_bar = px.bar(
    hourly_yearly,
    x='Hour',
    y='NormalizedCount',
    color='CrimeType',
    animation_frame='Year',
    barmode='group',
    title='Hourly Crime Distribution by Year (Animated Bar Chart)',
    labels={'NormalizedCount': 'Probability', 'Hour': 'Hour of Day', 'CrimeType': 'Crime Type'},
    range_x=[-0.5, 23.5],
    range_y=[0, 0.2]
)

# Start with all traces hidden — only on base traces, NOT on frames.
fig_bar.update_traces(visible='legendonly')

fig_bar.update_layout(
    legend=dict(
        title='Crime Type',
        orientation='v',
        yanchor='top', y=1,
        xanchor='left', x=1.02
    ),
    xaxis=dict(dtick=1),
    width=950,
    height=550,
    margin=dict(r=200)
)

fig_bar.show()
````

### Exercise 2.2 — Observations

**Yearly count check:** Some low-frequency crime types (e.g., Arson, Embezzlement, Suicide) have very small yearly counts, leading to noisy normalized distributions in individual years. This is expected — normalizing small counts amplifies random variation.

**Patterns over time:**
- **Stable patterns**: Assault, Larceny/Theft, and Robbery maintain remarkably consistent hourly shapes across most years. Assault consistently peaks late at night; Larceny/Theft peaks during afternoon hours.
- **Shifting patterns**: Drug-related and Liquor Laws offenses show more year-to-year variability, possibly reflecting changing enforcement priorities rather than changes in actual behavior.

**COVID effects (2020–2021):**
- Overall counts dropped for most crime types, but the *shapes* of hourly distributions shifted too. For example, daytime property crimes (Larceny/Theft) show a flatter distribution during lockdown periods, consistent with fewer people and businesses being active during normal hours.
- Some nighttime crimes (Assault, Robbery) show compressed peaks, likely reflecting reduced nightlife activity.

**Noisy distributions in recent years:**
- Low-count crime types like Arson, Suicide, and Embezzlement become visibly noisier as you advance through years — especially in years where total counts drop. This is a statistical artifact: fewer observations → more sampling noise in the normalized distribution.

**Line chart vs. bar chart:**
- The **line chart** makes temporal *shape* evolution easier to see — you can trace the contour of each distribution and notice when peaks shift or flatten.
- The **bar chart** (grouped) makes it easier to compare exact values at specific hours, but with 12 crime types × 24 hours, the bars become very thin and hard to read. The grouped layout is too crowded for this many categories.
- **Verdict**: The animated line chart is more effective for this data.

**Animation vs. small multiples:**
- **Animation** is good for casual exploration and presentations — it's engaging and lets you "watch" change unfold. But it's hard to compare non-adjacent years (e.g., 2005 vs. 2020) because you can't see both at once.
- **Small multiples** (one subplot per year) would be better for detailed analytical comparison — you can scan across all years simultaneously and spot trends or outliers at a glance.
- **Trade-off**: Animation favors storytelling and engagement; small multiples favor precision and comparison. For a research paper or policy report, small multiples would likely be more effective. For a presentation or dashboard, animation wins.
